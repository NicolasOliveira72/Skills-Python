```python
import pandas as pd
```

```python
vendas = {
    'Produto': ['Camisa', 'Calça', 'Saia', 'Camisa', 'Calça', 'Saia', 'Camiseta', 'Calça', 'Saia'],
    'Preço': [50, 100, 80, 55, 110, 85, 45, 95, 90],
    'Quantidade': [10, 15, 20, 5, 10, 15, 30, 20, 10],
    'Mês': ['Jan', 'Jan', 'Jan', 'Fev', 'Fev', 'Fev', 'Mar', 'Mar', 'Mar']
}
```

```python
vendas_DataFrame = pd.DataFrame(vendas)
vendas_DataFrame
```

```python
# Calculando a receita total:

vendas_DataFrame["Receita"] = vendas_DataFrame["Preço"] * vendas_DataFrame["Quantidade"]
vendas_DataFrame
```

```python
# Sumarizando a receita do produto:

agrupamento = vendas_DataFrame.groupby("Produto")["Receita"].sum().reset_index()
agrupamento
```

```python
import matplotlib.pyplot as pt
```

```python
pt.figure(figsize=(8,5))
pt.bar(agrupamento["Produto"], agrupamento["Receita"], color="red")
pt.title("Faturamento por produto")
pt.xlabel("Produto")
pt.ylabel("Receita", color="green")
pt.xticks(rotation=45)
```
