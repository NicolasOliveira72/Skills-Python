```python
dados_notas = {
    'Nome': ['Ana', 'Caio', 'Cesar', 'David', 'Eva'],
    'Matemática': [85, 90, 78, 88, 92],
    'Física': [80, 75, 88, 85, 95],
    'Química': [78, 82, 85, 80, 88]
}
```

```python
new_dados_notas = pd.DataFrame(dados_notas)
new_dados_notas
```

```python
# Calculando a média das notas:

new_dados_notas["Média"] = new_dados_notas[["Matemática", "Física", "Química"]].mean(axis=1)
new_dados_notas
```

```python
# Alunos com média maior que 82.5:

alunos_aprovados = new_dados_notas[new_dados_notas["Média"] >= 82.5]
alunos_aprovados
```

```python
# Média dos alunos aprovados:

new_dados_notas[["Nome", "Média"]]
```
