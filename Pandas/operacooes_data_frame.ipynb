```python
import pandas as pd
import matplotlib.pyplot as ptl
```

```python
data = {
    'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Idade': [25, 30, 35, 40, 27],
    'Cargo': ['Analista', 'Gerente', 'CEO', 'Analista', 'Coordenador'],
    'Salario': [5000, 8000, 15000, 4800, 6000]
}
```

```python
df = pd.DataFrame(data)
df
```
