```python
# Vendas mensais com Series:

vendas = pd.Series(
    [2000, 2300, 1800, 2020, 3020, 1500, 2480, 3060, 4050, 2830, 2130, 2350],
    index=["janeiro", "fevereiro", "março", "abril", "maio", "junho", 
    "julho", "agosto", "setembro", "outubro", "novembro", "dezembro"]
)
```

```python
vendas
```

```python
mediaVendas = int(vendas.mean())
```

```python
mediaVendas
```

```python
maiorVenda = vendas.idxmax()
```

```python
maiorVenda
```

```python
vendasAcimadaMedia = vendas[vendas > mediaVendas]
```

```python
vendasAcimadaMedia
```
